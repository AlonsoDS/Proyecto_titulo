```python
import json
import os
import pandas as pd
```

```python
# Ruta a la carpeta con los archivos JSON
ruta_carpeta = 'json_deptos_rm'

# Obtener todos los archivos JSON ordenados
archivos_json = sorted([f for f in os.listdir(ruta_carpeta) if f.endswith('.json')])

data = []

for idx, archivo in enumerate(archivos_json, 1):
    ruta_archivo = os.path.join(ruta_carpeta, archivo)
    with open(ruta_archivo, 'r', encoding='utf-8') as f:
        contenido = json.load(f)

    for feature in contenido.get('features', []):
        propiedades = feature.get('properties', {})
        geometria = feature.get('geometry', {})
        tilequery = propiedades.get("tilequery", {})

        fila = {
            'archivo_origen': archivo,
            'longitude': geometria.get('coordinates', [None, None])[0],
            'latitude': geometria.get('coordinates', [None, None])[1],
        }

        # Aplanar el resto de las propiedades directamente
        fila.update(propiedades)

        # Flatten tilequery
        fila['tilequery_distance'] = tilequery.get('distance', None)
        fila['tilequery_geometry'] = tilequery.get('geometry', None)

        data.append(fila)

# Crear DataFrame
df = pd.DataFrame(data)

# Exportar a CSV
df.to_csv('pois_viviendas.csv', index=False, encoding='utf-8')

print(f"Dataset exportado exitosamente con {len(df)} filas.")
```
