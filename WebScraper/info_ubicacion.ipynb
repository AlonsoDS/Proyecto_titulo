```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('dataset_completo.csv')
```

```python
conteo = df['region'].value_counts()

plt.figure(figsize=(10, 4)) 
conteo.head(20).plot(kind='bar')

plt.title('Frecuencia de valores nicos en la columna "region"')
plt.xlabel('Valor nico')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
df_rm = df[df['region'] == 'Regin Metropolitana de Santiago'].copy()
```

```python
conteo = df_rm['comuna'].value_counts()

plt.figure(figsize=(10, 4)) 
conteo.head(30).plot(kind='bar')

plt.title('Frecuencia de valores nicos en la columna "comuna"')
plt.xlabel('Valor nico')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```
