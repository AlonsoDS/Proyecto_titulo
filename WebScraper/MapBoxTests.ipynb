```python
import pandas as pd
import numpy as np
import requests
import json
pd.options.display.max_columns = None
```

```python
#dataset = pd.read_csv('scrap_depto/datos_deptos.csv')
df1 = pd.read_csv('scrap_depto/datos_deptos_link_largo.csv')
df2 = pd.read_csv('scrap_depto/datos_deptos.csv')
df = pd.concat([df1, df2], ignore_index=True)
df.head(5)
```

```python
df['link'] = df['link'].str[:45]
df.head(5)
```

```python
df['id'] = df['link'].str.extract(r'MLC-(\d+)')

# Mover la columna 'id' al principio
columnas = ['id'] + [col for col in df.columns if col != 'id']
df = df[columnas]
```

```python
df.sample(3)
```

```python
df[['latitud', 'longitud']] = df['coordenadas'].str.split('%2C', expand = True)
df['latitud'] = pd.to_numeric(df['latitud'])
df['longitud'] = pd.to_numeric(df['longitud'])
```

```python
df = df.drop(columns=['coordenadas'])
```

```python
df['monto'] = df['monto'].astype(str)
df.info()
```

```python

# Valor actual de la UF
valor_uf = 39036.45

# Paso 1: Filtrar filas con divisa en "$"
mask_pesos = df['divisa'] == "$"

# Paso 2: Limpiar la columna 'monto' removiendo puntos y convirtiendo a float
df.loc[mask_pesos, 'monto'] = (
    df.loc[mask_pesos, 'monto']
    .str.replace('.', '', regex=False)  # Eliminar separador de miles
    .str.replace(',', '.', regex=False)  # Por si hay coma decimal
    .astype(float)
)

# Paso 3: Convertir los montos a UF
df.loc[mask_pesos, 'monto'] = df.loc[mask_pesos, 'monto'] / valor_uf

# Paso 4: Cambiar la divisa a "UF"
df.loc[mask_pesos, 'divisa'] = "UF"
```

```python
columnas_y_unidades = {
    'monto': '',            # Ya lo tratamos antes, solo asegúrate que esté en float
    'ppm2': ' UF/m²',
    'ppm2_avg': ' UF/m²',
    'superficie_t': ' m²',
    'superficie_u': ' m²',
    'terraza': ' m²'
}

# Limpiar y convertir a float
for col, unidad in columnas_y_unidades.items():
    df[col] = (
        df[col]
        .astype(str)                          # Asegurar que sean strings
        .str.replace('.', '', regex=False)    # Eliminar separador de miles
        .str.replace(',', '.', regex=False)   # Reemplazar coma por punto decimal
        .str.replace(unidad, '', regex=False) # Eliminar unidad
        .str.strip()                          # Eliminar espacios al inicio/fin
        .replace('', '0')                     # Por si quedó vacío
        .astype(float)
    )
```

```python
import requests

def get_location_details(latitude, longitude, mapbox_token):
    url = f"https://api.mapbox.com/geocoding/v5/mapbox.places/{longitude},{latitude}.json"
    
    params = {
        'access_token': mapbox_token,
        'types': 'address,place,region,country',
        'language': 'es'
    }

    response = requests.get(url, params=params)
    data = response.json()

    # Diccionario para guardar resultados
    result = {
        'calle': None,
        'numero': None,
        'comuna': None,
        'region': None,
        'pais': None
    }

    if 'features' in data:
        for feature in data['features']:
            if 'place_type' not in feature:
                continue

            tipo = feature['place_type'][0]

            if tipo == 'address':
                result['calle'] = feature.get('text')
                result['numero'] = feature.get('address')
            elif tipo == 'place':
                result['comuna'] = feature.get('text')
            elif tipo == 'region':
                result['region'] = feature.get('text')
            elif tipo == 'country':
                result['pais'] = feature.get('text')

    return result

# Coordenadas de Santiago Centro -70.59675112366676, Latitud = -33.45219353855461
latitude = -33.45219353855461
longitude = -70.59675112366676

mapbox_token = "pk.eyJ1IjoiYWxvbnNvZHMiLCJhIjoiY200MDhjOG11MHo0eTJtcTRuN2o5OGE5YSJ9.e9NX177oozaJ1NBerR2sTg"  # Reemplaza con tu token de Mapbox

detalles = get_location_details(latitude, longitude, mapbox_token)
print(detalles)
```

```python
import pandas as pd
import time
import random
import requests

def obtener_direccion_desde_coordenadas(lat, lon, token):
    url = f"https://api.mapbox.com/geocoding/v5/mapbox.places/{lon},{lat}.json"
    params = {
        'access_token': token,
        'types': 'address,place,region,country',
        'language': 'es'
    }
    response = requests.get(url, params=params)
    if response.status_code != 200:
        return {'pais': None, 'region': None, 'comuna': None, 'calle': None, 'direccion_num': None}

    data = response.json()
    features = data.get('features', [])

    pais = region = comuna = calle = direccion_num = None

    for f in features:
        if 'country' in f['place_type']:
            pais = f['text']
        elif 'region' in f['place_type']:
            region = f['text']
        elif 'place' in f['place_type']:
            comuna = f['text']
        elif 'address' in f['place_type']:
            calle = f['text']
            # Número no siempre está separado, dejamos como calle por ahora
            direccion_num = calle

    return {'pais': pais, 'region': region, 'comuna': comuna, 'calle': calle, 'direccion_num': direccion_num}

def procesar_batch(input_csv, output_csv, mapbox_token):
    df = pd.read_csv(input_csv)

    paises = []
    regiones = []
    comunas = []
    calles = []
    direcciones = []

    for idx, row in df.iterrows():
        lat = row['latitud']
        lon = row['longitud']

        datos = obtener_direccion_desde_coordenadas(lat, lon, mapbox_token)

        paises.append(datos['pais'])
        regiones.append(datos['region'])
        comunas.append(datos['comuna'])
        calles.append(datos['calle'])
        direcciones.append(datos['direccion_num'])

        # Pausa para no sobrepasar límite
        time.sleep(random.uniform(0.2, 0.3))

        if (idx + 1) % 500 == 0:
            print(f"{idx + 1} filas procesadas")

    df['pais'] = paises
    df['region'] = regiones
    df['comuna'] = comunas
    df['calle'] = calles
    df['direccion_num'] = direcciones

    df.to_csv(output_csv, index=False)
    print(f"Archivo exportado como {output_csv}")
```
