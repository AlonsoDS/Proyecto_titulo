```python

# Lista de columnas a mantener
columns_to_keep = [
    "longitude", "latitude", "type", "filterrank",
    "category_en", "sizerank", "maki", "class", "name"
]

# Crear nuevo DataFrame con solo las columnas necesarias
df_clean = df_poi[columns_to_keep].copy()

# Eliminar duplicados (POIs únicos según estas columnas)
POIs_unique = df_clean.drop_duplicates().reset_index(drop=True)
```

```python
POIs_unique.info()
```

```python
POIs_unique.to_csv('pois_unique.csv', index=False, encoding='utf-8')
```

```python
import folium

# Centro aproximado de Santiago de Chile (puedes ajustarlo si quieres)
centro_santiago = [-33.45, -70.66]

# Crear el mapa
mapa = folium.Map(location=centro_santiago, zoom_start=12)

# Agregar los POIs al mapa
for _, row in POIs_unique.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=3,
        color='blue',
        fill=True,
        fill_opacity=0.6,
        popup=row["name"]
    ).add_to(mapa)

# Mostrar el mapa
mapa
```
