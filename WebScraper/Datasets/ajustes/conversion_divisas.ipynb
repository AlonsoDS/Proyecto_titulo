```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
```

```python
df= pd.read_csv('deptos_divisa_mixta.csv')
```

```python
df.info()
```

```python
df['divisa'].unique()
```

```python
# Valor actual de la UF
valor_uf = 39036.45

# Filtrar filas con divisa "$"
mask_pesos = df['divisa'] == "$"

# Convertir montos de pesos a UF (simple división)
df.loc[mask_pesos, 'monto'] = df.loc[mask_pesos, 'monto'] / valor_uf

# Actualizar la divisa a "UF"
df.loc[mask_pesos, 'divisa'] = "UF"
```

```python
df['monto'].describe()
```

```python
df.sample(10)
```

```python
df['monto'] = df['monto'].round(0).astype(int)
```

```python
df = df.drop(columns=['Unnamed: 0'])
```
