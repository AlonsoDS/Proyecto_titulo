```python
df = pd.read_csv('Datasets\deptos_en_limpieza.csv')
#df_dropna = pd.read_csv('Datasets\Dataset_sin_nan_9803.csv')
```

```python
# Centro aproximado de Santiago de Chile (puedes ajustarlo si quieres)
centro_santiago = [-33.45, -70.66]

# Crear el mapa
mapa = folium.Map(location=centro_santiago, zoom_start=12)

# Agregar los POIs al mapa
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitud"], row["longitud"]],
        radius=3,
        color='blue',
        fill=True,
        fill_opacity=0.6,
        popup=row["id"]
    ).add_to(mapa)

# Mostrar el mapa
mapa
```

```python
df.drop(columns=["ppm2","cantidad_pisos", "piso", "orientacion_norm", 'superficie_u', 'tipo_depto', 'tipo_depto_norm'], inplace=True)
```

```python
column_order = [
    'id',
    'monto',
    'superficie_t',
    # Características estructurales/comodidad y sus flags
    'dormitorios',
    'dormitorios_faltante',
    'banos',
    'banos_faltante',
    'antiguedad',
    'antiguedad_faltante',
    # Flags de Orientación (la columna original 'orientacion_norm' probablemente ya no es necesaria)
    'Or_N',
    'Or_S',
    'Or_E',
    'Or_O',
    'Or_Faltante',
    'terraza',
    'estacionamiento',
    'bodegas',
    # Flags de Tipo de Departamento
    'flag_Departamento',
    'flag_Multinivel', # Asumo que es 'flag_Multinivel'
    'flag_Semipiso',
    'flag_Premium',    # Asumo que es 'flag_Premium'
    'flag_Monoambiente',
    'flag_Loft',
    # Variables Geográficas
    'latitud',
    'longitud',
    'comuna',
    'calle'
]

# Verificar si todas las columnas deseadas existen en el DataFrame
# Esto es importante para evitar errores si alguna columna fue dropeada o tiene un nombre diferente
existing_columns = df.columns.tolist()
final_column_order = [col for col in column_order if col in existing_columns]


# Reordenar el DataFrame
df = df[final_column_order]

print("Columnas reordenadas exitosamente. Nuevo orden de columnas:")
print(df.columns.tolist())
```
