```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv('dataset_vcr_compact.csv')
```

```python
# Centro aproximado de Santiago de Chile 
centro_santiago = [-33.45, -70.66]

# Crear el mapa
mapa = folium.Map(location=centro_santiago, zoom_start=12)

# Agregar los deptos al mapa
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitud"], row["longitud"]],
        radius=1,
        color='blue',
        fill=True,
        fill_opacity=0.6,
        popup=row["latitud"]
    ).add_to(mapa)

# Guardar y abrir en el navegador
mapa_path = "mapa_santiago_deptos.html"
mapa.save(mapa_path)
```
